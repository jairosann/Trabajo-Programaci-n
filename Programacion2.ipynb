```python
def resumen_columna(df, col):
    serie = df[col]

    print(f"=== Resumen de '{col}' ===")
    print("Tipo:", serie.dtype)
    print("N nulos:", serie.isna().sum())
    
    # Si es numérica, imprime estadísticas numéricas
    if pd.api.types.is_numeric_dtype(serie):
        print("Media:", serie.mean())
        print("Mediana:", serie.median())
        print("Moda:", list(serie.mode()))
        print("Desviación típica:", serie.std())
        print("Mínimo:", serie.min())
        print("Máximo:", serie.max())
        print("Percentiles (25-50-75):")
        print(serie.quantile([0.25, 0.5, 0.75]))
    else:
        # Si es categórica o string
        print("Valores únicos:", serie.nunique())
        print("Moda:", list(serie.mode()))
        print("Frecuencias:")
        print(serie.value_counts())


columna = 'Age'

resumen_columna(Bank_Customers, columna)
```

# Función de análisis por columna y graficamiento

Este codigo es para analizar cada columna por un lado y luego tras el analisis de la columna ver si nos interesa mas imputar por la media o la mediana.


```python
def analizar_columna(df, col):
    mu = df[col].mean()
    sigma = df[col].std()
    low = mu - 2*sigma
    high = mu + 2*sigma
    
    print(f"=== {col} ===")
    print("Media:", mu)
    print("Mediana:", df[col].median())
    print("Desv. típica:", sigma)
    print("Límites ±2σ:", low, high)
    
    # Gráfica
    plt.figure(figsize=(8,4))
    plt.hist(df[col], bins=30, edgecolor="black")
    plt.axvline(low, color='red', linestyle='--')
    plt.axvline(high, color='red', linestyle='--')
    plt.axvline(mu, color='green')
    plt.title(f"Distribución de {col} con límites ±2σ")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()

def imputar_media(df, col):
    mu = df[col].mean()
    sigma = df[col].std()
    low = mu - 2*sigma
    high = mu + 2*sigma
    df.loc[(df[col] < low) | (df[col] > high), col] = int(mu)

def imputar_mediana(df, col):
    med = df[col].median()
    sigma = df[col].std()
    mu = df[col].mean()
    low = mu - 2*sigma
    high = mu + 2*sigma
    df.loc[(df[col] < low) | (df[col] > high), col] = med
```

# Función de analísis estadístico por columnas

```python
import numpy as np
import matplotlib.pyplot as plt

def analizar_columna(df, columna):
    """
    Genera histograma y boxplot de una columna numérica del DataFrame,
    e imprime estadísticas descriptivas clave y outliers según IQR.
    """
    datos = df[columna].dropna()

    # ---------- Estadísticos básicos ----------
    n_total = len(df[columna])
    n_validos = len(datos)
    n_nan = n_total - n_validos

    media = datos.mean()
    mediana = datos.median()
    # Moda: puede devolver varias; nos quedamos con la primera
    moda = datos.mode()
    moda = moda.iloc[0] if len(moda) > 0 else np.nan

    std = datos.std()
    minimo = datos.min()
    maximo = datos.max()
    skew = datos.skew()

    Q1 = datos.quantile(0.25)
    Q3 = datos.quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    outliers = datos[(datos < limite_inf) | (datos > limite_sup)]

    # ---------- Gráficas: histograma + boxplot ----------
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Histograma
    axes[0].hist(datos, bins=20, edgecolor="black")
    axes[0].axvline(media, color="red", linestyle="--", label=f"Media = {media:.2f}")
    axes[0].axvline(mediana, color="green", linestyle="--", label=f"Mediana = {mediana:.2f}")
    axes[0].set_title(f"Histograma de {columna}")
    axes[0].set_xlabel(columna)
    axes[0].set_ylabel("Frecuencia")
    axes[0].legend()

    # Boxplot
    axes[1].boxplot(datos, vert=True, showfliers=True)
    axes[1].set_title(f"Boxplot de {columna}")
    axes[1].set_ylabel(columna)

    plt.tight_layout()
    plt.show()

    # ---------- Resumen numérico ----------
    print(f"\nAnálisis estadístico para: {columna}")
    print("-----------------------------------------")
    print(f"Nº total de registros: {n_total}")
    print(f"Nº valores válidos:    {n_validos}")
    print(f"Nº valores NaN:        {n_nan}")
    print()
    print(f"Media:           {media:.4f}")
    print(f"Mediana:         {mediana:.4f}")
    print(f"Moda:            {moda:.4f}")
    print(f"Desviación típ.: {std:.4f}")
    print(f"Mínimo:          {minimo}")
    print(f"Máximo:          {maximo}")
    print()
    print(f"Q1 (25%):        {Q1:.4f}")
    print(f"Q3 (75%):        {Q3:.4f}")
    print(f"IQR:             {IQR:.4f}")
    print(f"Skew (sesgo):    {skew:.4f}")
    print()
    print(f"Outliers (IQR):  {len(outliers)} valores")
    print(f" - Límite inf.:  {limite_inf:.4f}")
    print(f" - Límite sup.:  {limite_sup:.4f}")
    print("-----------------------------------------\n")


def analizar_numericas(df):
    """
    Aplica analizar_columna a todas las columnas numéricas del DataFrame.
    """
    columnas_num = df.select_dtypes(include=["int64", "float64"]).columns
    for col in columnas_num:
        print(f"===== {col} =====")
        analizar_columna(df, col)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_relacion(df, col_x, col_y, sample_frac=1.0):
    """
    Muestra la relación entre dos columnas mediante:
    - nube de puntos (X vs Y)
    - recta de regresión lineal
    - coeficiente de correlación de Pearson en el título

    sample_frac: fracción de filas a usar (por si tienes muchas, ej. 0.3).
    """

    # Seleccionamos columnas y quitamos NaN
    datos = df[[col_x, col_y]].dropna()

    # Muestreo opcional para no saturar la gráfica
    if sample_frac < 1.0:
        datos = datos.sample(frac=sample_frac, random_state=0)

    x = datos[col_x].values
    y = datos[col_y].values

    # Correlación de Pearson
    r = np.corrcoef(x, y)[0, 1]

    plt.figure(figsize=(7, 5))

    # Nube de puntos
    plt.scatter(x, y, alpha=0.3, s=10, label="Datos")

    # Recta de regresión lineal
    coef = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min(), x.max(), 100)
    y_line = np.poly1d(coef)(x_line)
    plt.plot(x_line, y_line, linewidth=2, label="Tendencia lineal")

    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title(f"Relación {col_x} vs {col_y} (r = {r:.2f})")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_relacion(Bank_Customers, "Credit_Limit", "Avg_OpenToBuy")
plot_relacion(Bank_Customers, "Total_Trans_Amt", "Total_Trans_Ct")
plot_relacion(Bank_Customers, "Avg_Utilization_Ratio", "Avg_OpenToBuy")
plot_relacion(Bank_Customers, "Avg_Utilization_Ratio", "Credit_Limit")

```

```python
import pandas as pd

df = pd.read_csv("Bank_Customers_mod2.csv")

# Normalizadores
max_trans = df["Trans_Count"].max()
max_inactive = df["Inactive_Months"].max()
max_products = df["Products"].max()
max_contacts = df["Bank_Contacts"].max()

scores = []

for _, fila in df.iterrows():
    r_inactive = fila["Inactive_Months"] / max_inactive
    r_trans = 1 - (fila["Trans_Count"] / max_trans)
    r_products = 1 - (fila["Products"] / max_products)
    r_contacts = 1 - (fila["Bank_Contacts"] / max_contacts)

    score = (
        0.35 * r_inactive +
        0.35 * r_trans +
        0.20 * r_products +
        0.10 * r_contacts
    )

    scores.append(score)

df["Riesgo_Score"] = (pd.Series(scores) * 100).round(2)

# Percentiles para separar grupos
p40 = df["Riesgo_Score"].quantile(0.40)
p80 = df["Riesgo_Score"].quantile(0.80)

def clasificar(score):
    if score <= p40:
        return "Bajo"
    elif score <= p80:
        return "Medio"
    else:
        return "Alto"

df["Riesgo_Abandono"] = df["Riesgo_Score"].apply(clasificar)

df.head()
```
